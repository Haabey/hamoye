```python
# Import the necessary python libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
# Load the dataset

data = pd.read_csv("/content/Dataset/energydata_complete.csv")
```

```python


# Extract relevant columns
X = data["T2"].values.reshape(-1, 1)
y = data["T6"].values

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the temperature inside based on outside temperature
y_pred = model.predict(X)

# Calculate the RMSE
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

# Round and format the RMSE to three decimal places
rmse_rounded = round(rmse, 3)

# Print the RMSE
print(f"Root Mean Squared Error: {rmse_rounded}")
```

```python
# Check if "date" and "lights" columns exist in the data DataFrame
if "date" in data.columns and "lights" in data.columns:
    print("Both 'date' and 'lights' columns exist in the DataFrame.")
else:
    print("One or both of 'date' and 'lights' columns are missing from the DataFrame.")
```

```python
# Preprocess data

columns_to_drop = ["T2", "T6", "Appliances"]
X = data.drop(columns=columns_to_drop)
y = data["Appliances"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Normalise the data
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)
```

```python
X.head()
```

```python
# Create and fit the model
model = LinearRegression()
model.fit(X_train_normalized, y_train)
```

```python
# Calculate the mean absolute error on the training set
y_pred_train = model.predict(X_train_normalized)
mae_train = mean_absolute_error(y_train, y_pred_train)

# Round and format the MAE to three decimal places
mae_train_rounded = round(mae_train, 3)

# Print the MAE
print("Mean Absolute Error on the training set:", mae_train_rounded)
```

```python

y_pred = model.predict(X_test_normalized)
r2_test = r2_score(y_test, y_pred)
print(f"R^2 on test set: {r2_test:.3f}")
```

```python
from sklearn.metrics import mean_squared_error

n = X_test_normalized.shape[0]
p = X_test_normalized.shape[1]

adjusted_r2 = 1 - ((1 - r2_test) * (n - 1) / (n - p - 1))
print(f"Adjusted R^2 on test set: {adjusted_r2:.3f}")
```

```python
# Training a ridge regression model with default parameters

ridge_model = Ridge()
ridge_model.fit(X_train, y_train)

# Predict on the test set
y_pred_ridge = ridge_model.predict(X_test)

# Calculate the RMSE for the test set
from sklearn.metrics import mean_squared_error

rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))

# Round and format the RMSE to three decimal places
rmse_ridge_rounded = round(rmse_ridge, 3)

# Print the RMSE
print(f"Root Mean Squared Error (Ridge Regression): {rmse_ridge_rounded}")

# Compare RMSE with the previous model
print("Change in RMSE:", rmse_ridge_rounded - rmse_rounded)
```

```python
# Train a Lasso regression model with default parameters

lasso_model = Lasso()
lasso_model.fit(X_train, y_train)

# Obtain the new feature weights
lasso_coef = lasso_model.coef_

# Count the number of non-zero feature weights
non_zero_features = np.count_nonzero(lasso_coef)

# Print the results
print(f"Number of non-zero features: {non_zero_features}")

# Print the feature weights
print("Feature weights:", lasso_coef)
```

```python
# Predict on the test set
y_pred_lasso = lasso_model.predict(X_test)

# Calculate the RMSE for the test set
from sklearn.metrics import mean_squared_error

rmse_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso))

# Round and format the RMSE to three decimal places
rmse_lasso_rounded = round(rmse_lasso, 3)

# Print the RMSE
print(f"Root Mean Squared Error (Lasso Regression): {rmse_lasso_rounded}")
```
